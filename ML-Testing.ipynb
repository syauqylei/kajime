```python
import os
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(f"{os.getcwd()}/datasets/Salary_Data.csv")
x = df.to_numpy()[:,0]
y = df.to_numpy()[:,1]
```

```python
plt.plot(x,y, 'g^')
```

```python
import numpy as np
from src.solvers.gradient_descent import GradientDescent

gdSolver = GradientDescent(eta=0.01,max_iter=2500)

coefs = gdSolver.solve(y, x)

x_ = np.linspace(0,15,num=100)
y_ = np.array([ coefs[0] + coefs[1] * xi for xi in x_ ])
plt.plot(x,y, 'g^', x_,y_)
```

```python
from src.regression import LinearRegression

linreg = LinearRegression(eta=0.01,epoch=2500)
linreg.fit(x,y)

y_pred = linreg.predict(x)

plt.plot(x,y, 'g^', x,y_pred)
```
